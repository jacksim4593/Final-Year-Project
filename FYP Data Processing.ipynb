```python
import pandas as pd
import numpy as np
from math import radians, cos, sin, asin, sqrt
from geopy.distance import great_circle

#This block of code remove the road that have traffic incidents in it. The condition is within 5mins and within a radius of 500m.
def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    return c * 6371 * 1000

def within_threshold(lat1, lon1, lat2, lon2, threshold):
    return haversine(lat1, lon1, lat2, lon2) <= threshold

def time_difference_within_5_minutes(t1, t2):
    delta = abs(pd.to_datetime(t1) - pd.to_datetime(t2))
    return delta <= pd.Timedelta(minutes=5)

# Convert the Location column to a list of lists of coordinates (lat, lon) tuples
filteredspeed_df['Coordinates'] = filteredspeed_df['Location'].apply(lambda x: [tuple(map(float, coord.split())) for coord in x.split(', ')])

# Pre-filtering: Select only rows where dates match
dates_match = filteredspeed_df['date'].isin(filteredtraffic_df['date'])
pre_filtered_speed = filteredspeed_df[dates_match]

# Perform the optimized filtering
filtered_indexes = set()
for idx, row in filteredtraffic_df.iterrows():
    matching_speed_rows = pre_filtered_speed[
        (pre_filtered_speed['date'] == row['date']) &
        (pre_filtered_speed['Time'].apply(lambda x: time_difference_within_5_minutes(x, row['Time'])))
    ]
    for speed_idx, speed_row in matching_speed_rows.iterrows():
        if any(within_threshold(row['Latitude'], row['Longitude'], coord[0], coord[1], 500) for coord in speed_row['Coordinates']):
            filtered_indexes.add(speed_idx)

# Remove the matched rows from the filteredspeed_df
filteredspeed_df = filteredspeed_df.drop(filtered_indexes)

# Update the Coordinates column to include both starting and ending coordinates
filteredspeed_df.loc[:, 'Coordinates'] = filteredspeed_df['Location'].apply(lambda x: [tuple(map(float, coord.split())) for coord in x.split(', ')])

filteredspeed_df
```

```python
import os
import json
import pandas as pd
from datetime import datetime

folder_path = r"D:\OneDrive\OneDrive - Nanyang Technological University\Desktop\NTU Student Life\Year 4\FYP\Rainfall Data"

def parse_rainfall_data(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    
    stations = data["metadata"]["stations"]
    timestamp_list = []
    station_data_list = []
    
    for item in data['items']:
        timestamp = datetime.fromisoformat(item['timestamp'].replace('+08:00', ''))
        for reading in item['readings']:
            station_data = reading.copy()
            station_data['timestamp'] = timestamp
            
            # Add station details to the station_data
            for station in stations:
                if station["id"] == station_data["station_id"]:
                    station_data["name"] = station["name"]
                    station_data["latitude"] = station["location"]["latitude"]
                    station_data["longitude"] = station["location"]["longitude"]
                    break
            
            station_data_list.append(station_data)
    
    return station_data_list

all_rainfall_data = []
for file_name in os.listdir(folder_path):
    if not file_name.endswith(".txt") and not file_name.endswith(".json"):
        file_path = os.path.join(folder_path, file_name)
        all_rainfall_data.extend(parse_rainfall_data(file_path))

rainfall_df = pd.DataFrame(all_rainfall_data)
rainfall_df
```

```python
# Convert the timestamp column to a pandas datetime object
rainfall_df['timestamp'] = pd.to_datetime(rainfall_df['timestamp'])

# Filter the DataFrame to only include rows between 6 AM and 8 AM
filtered_rainfall_df = rainfall_df[(rainfall_df['timestamp'].dt.hour >= 6) & (rainfall_df['timestamp'].dt.hour < 8)]

# Display the filtered DataFrame
filtered_rainfall_df
```

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import BallTree

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    dLat = np.radians(lat2 - lat1)
    dLon = np.radians(lon2 - lon1)
    a = np.sin(dLat/2) * np.sin(dLat/2) + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dLon/2) * np.sin(dLon/2)
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    d = R * c
    return d

# Prepare the data
start_locations = np.array([np.array(loc, dtype=float) for loc in filteredspeed_df['start_location'].values.tolist()])
end_locations = np.array([np.array(loc, dtype=float) for loc in filteredspeed_df['end_location'].values.tolist()])
rainfall_locations = np.array([eval(loc) for loc in filtered_rainfall_df['Location'].values], dtype=float)

# Create a BallTree using the haversine metric
tree = BallTree(np.radians(rainfall_locations), metric='haversine')

# Calculate the nearest neighbor for start and end locations
start_indices = tree.query(np.radians(start_locations), return_distance=False)
end_indices = tree.query(np.radians(end_locations), return_distance=False)

# Flatten the indices arrays
start_indices = start_indices.flatten()
end_indices = end_indices.flatten()

# Find the nearest station for each pair of start and end locations
start_distances = np.array([haversine(start_locations[i, 0], start_locations[i, 1], rainfall_locations[start_indices[i], 0], rainfall_locations[start_indices[i], 1]) for i in range(start_locations.shape[0])])
end_distances = np.array([haversine(end_locations[i, 0], end_locations[i, 1], rainfall_locations[end_indices[i], 0], rainfall_locations[end_indices[i], 1]) for i in range(end_locations.shape[0])])
nearest_station_indices = np.where(start_distances < end_distances, start_indices, end_indices)

# Add the 'nearest_station' column to the filteredspeed_df DataFrame
filteredspeed_df['nearestrainfall_station'] = filtered_rainfall_df.iloc[nearest_station_indices]['station_id'].values

filteredspeed_df
```
